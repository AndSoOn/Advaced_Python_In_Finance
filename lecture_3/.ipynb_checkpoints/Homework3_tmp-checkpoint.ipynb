## Problem 1
$\renewcommand{bs}{\boldsymbol}$

We discussed three different ways of solving the least square problem of $\min_{\boldsymbol \beta}\Vert X \bs \beta - \boldsymbol y\Vert_2$ in the lecture: 1) the pseudo-inverse $X^+$ 2) QR decomposition 3) SVD. 

1. if $X$ is not fully ranked, i.e., its column vectors are linearly dependent, do the three methods still work?
2. if not, can you propose an modifications to the QR method, so that you can find the minimum value of $\Vert X \bs \beta - \boldsymbol y\Vert_2$ even when $X$ is rank deficient? Note that $\bs \beta$ is not unique when $X$ is not fully ranked, but there is still a deterministic answer to the minimum value of $\Vert X \bs \beta - \boldsymbol y\Vert_2$. 
3. what is an easy and effective method to find a $\bs \beta$ that is good enough when $A$ is not fully ranked in practice?

Hint:
- It is also possible to make SVD work with rank deficient matrix. Please refer to http://www.cs.cornell.edu/Courses/cs3220/2010sp/notes/svd.pdf if you are interested.

## Solution:
#### 1. 
The pseudo-inverse still works, as well as the SVD with some modifications. But QR decomposition no longer works. If $X$ is not fully ranked, then we will get something like $X = QR = Q \left[ \begin{array} \\ R_{1} & R_{2} \\ 0 & 0  \end{array}\right]$ where $R$ is not invertable.

#### 2.
We can introduce a permutation matrix $P$ such that $XP=QR, R = \left[ \begin{array} \\ R_{1} & R_{2} \\ 0 & 0  \end{array}\right]$  
Then $\Vert X \bs \beta - \boldsymbol y\Vert_2 = \Vert Q^{T}(XPP^{T}\bs \beta - y)\Vert_2 = \left \Vert \left( \begin{array} \\ R_{1}R_{2} \\ 0  \end{array}\right)P^{T}\bs \beta - Q^{T}\bs y \right \Vert_2$  

Which can be further rearranged as $\Vert R_1 y_1 + R_2 y_2 - c_1\Vert_2 + \Vert c_2\Vert_2 + \Vert d \Vert_2$  

Then we can simply minimize $\Vert R_1 y_1 + R_2 y_2 - c_1\Vert_2 + \Vert c_2\Vert_2 + \Vert d \Vert_2$

#### 3. 
Ridge regression is often the answer for this problem in practice, when $X$ is rank deficient.  
Besides, we can estimate $\bs \beta$ by implement lasso regression.

## Problem 2

Take the covariance matrix $V$ estimated from the CMT historical levels, compute the following:
1. its L-2 condition number
2. its singular values, and the ratio between the largest and smallest singular values
3. Comment on whether there could be troubled area when computing $\bs y = V^{-1} \bs x$ from this co-variance matrix, if so, give an example of $\bs x$ that the resulting $\bs y$ changes a lot even with a small perturbations in $x$.

## Solution:
#### 1. 

```python
from numpy.linalg import cond, svd

cmt_cov = cmt_rates.cov()
cond_l2 = cond(cmt_cov, p=2)

print "L-2 condition number is: {}".format(cond_l2)
```

#### 2.

```python
from numpy.linalg import svd
U, s, V = svd(cmt_cov)
print "Singular values are: "
print s
print "Ratio between the largest and smallest is: {}".format(s[0] / s[-1])
```

#### 3.

It is very likely to get into trouble when computing $\bs y = V^{-1} \bs x$ from this co-variance matrix since the condition number is considerable large. Below is an example:

```python
from numpy.random import random

cov_inv = np.linalg.inv(cmt_cov)
x = V[0, :]
numpy.random.seed(233)
x_pert = x + random(9) * 0.01
y = cov_inv.dot(x)
y_pert = cov_inv.dot(x_pert)

print "Original:   ", "  After Perturbations: "
for i in range(len(y)):
    print y[i],  y_pert[i]
```

#### 1.
The covariance of the daily interest rate **changes**:

```python
ir_diff = cmt_rates.diff()[1:]
cov = ir_diff.cov()
cov
```

#### 2.
First, two matrices give different results:

```python
corr = ir_diff.corr()
w_cov, v_cov = np.linalg.eig(cov)
w_corr, v_corr = np.linalg.eig(corr)
PCA_cov = 100 * np.cumsum(w_cov) / sum(w_cov)
PCA_corr = 100 * np.cumsum(w_corr) / sum(w_corr)

print "PCA from covariance:"
pd.set_option('display.precision', 3)
fmt.displayDF(pd.DataFrame({'P/C':range(1, len(PCA_cov)+1), 
                            'Eigenvalues':w_cov, 'Cumulative Var(%)':
                            PCA_cov}).set_index(['P/C']).T, "6f")
print "PCA from correlation:"
pd.set_option('display.precision', 3)
fmt.displayDF(pd.DataFrame({'P/C':range(1, len(PCA_corr)+1), 
                            'Eigenvalues':w_corr, 'Cumulative Var(%)':
                            PCA_corr}).set_index(['P/C']).T, "6f")
```

Then, taken volume factors into consideration, eigen vectors are still distictly different:

```python
factors = np.power(np.diag(cov), 0.5)
v_corr_adj = []
for i in range(9):
    v_corr_adj.append(v_corr.T[i] * factors)
v_corr_adj = np.array(v_corr_adj).T

for i in range(9):
    print v_cov[i, :] / v_corr_adj[i, :]
```

#### 6.
The re-estimated covariance matrix is as follows:

```python
ir_diff = cmt_rates.diff()[1:]
cov = ir_diff.cov()
num_pcs, num_iter, t = 3, 100, 250 * 20
w_cov, v_cov = np.linalg.eig(cov)
factors = np.power(w_cov[:num_pcs], 0.5)
pcs = np.array([v_cov[:, :num_pcs].T[i] * factors[i] for i in xrange(num_pcs)]).T    

ir_sim = []
for i in tqdm(xrange(num_iter)):
    tmp = []
    for j in xrange(t):
        rand = np.random.normal(size=num_pcs)
        val = pcs.dot(rand)
        tmp.append(val)
    ir_sim.append(tmp)
ir_sim_df = pd.DataFrame(np.vstack(ir_sim), columns=cmt_rates.columns)
ir_sim_cov = ir_sim_df.cov()
fmt.displayDFs(ir_sim_cov, headers=['Simulated'])
```

```python
fmt.displayDFs(cov, headers=['Original'])
```

For the comparison, we can creat a logical matrix which will be true for each element if the relative difference between them is less than 10 percent. 

We can see the two matrices are very similar.

```python
error = np.divide(np.subtract(cov, ir_sim_cov), cov)
TF = abs(error) < 0.1
fmt.displayDFs(TF, headers=['Logical Matrix'])
```

#### 7.
Assume the starting interest rate is 2%. 

```python
a = np.array(ir_sim)
cum = np.cumsum(a, 1) + 2
y1, y10 = cum[:, :, 2], cum[:, :, -2]

mu_1, mu_10 = np.mean(y1, 0), np.mean(y10, 0)
std_1, std_10 = np.std(y1, 0), np.std(y10, 0)
qt2_1, qt2_10 = np.percentile(y1, 2, axis=0), np.percentile(y10, 2, axis=0)
qt98_1, qt98_10 = np.percentile(y1, 98, axis=0), np.percentile(y10, 98, axis=0)

df1 = pd.DataFrame(np.array([mu_1, std_1, qt2_1, qt98_1]).T, 
                   index=np.arange(len(mu_1))/250.,
                  columns=["mean", "std", "2%", "98%"])
df1.plot(figsize=(10, 5), title="1Y rates")
```

```python
df10 = pd.DataFrame(np.array([mu_10, std_10, qt2_10, qt98_10]).T, 
                   index=np.arange(len(mu_10))/250.,
                  columns=["mean", "std", "2%", "98%"])
df10.plot(figsize=(10, 5), title="10Y rates")
```
