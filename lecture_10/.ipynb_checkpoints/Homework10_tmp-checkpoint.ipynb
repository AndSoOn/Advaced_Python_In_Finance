## Problem 1

Use the dual objective discussed in class to solve the following maximum entropy optimization problems.

Find the the maximum entropy (ME) distribution without any prior distributions (equivalent to uniform priors) with the following constraints:

1. with constraints of $\mathbb{E}[x] = 0$ and $\mathbb{E}[x^2] = 1$; show that the resulting distribution is indeed normal.
2. does your ME distribution from #1 depend on the range of the discrete samples? for example, do the resulting ME distribution look normal if the discrete samples span only [-1.5, 1.5]? explain why.
3. define a transformation of $y = e^x$, where $y$ is a lognormal distribution if $x$ is normal. Solve the ME problem using equal spaced samples in $y$ with the constraint $\mathbb{E}[\log(y)]=0$ and $\mathbb{E}[\log^2(y)] = 1$, and check if the resulting $y$ distribution is lognormal (it is not). 
4. Explain why your step 3 do not result in a lognormal distribution.
5. [optional] Can you adjust the prior distribution in #3 to reproduce the similar results as in #1 without changing the equal-spaced sampling in $y$?

## Solution:
#### 1.
As can be seen from the Q-Q plot, the distribution under the given constrains are essentially normal with mean 0 and var 1.

```python
x = np.arange(-10, 10, .001)
a = np.array([x, x*x])
u = np.array([0., 1.])
e = np.array([0., 0.])
q = np.ones(np.size(x))/len(x)
dual = me.MaxEntDual(q, a, u, e)

res = minimize(dual.dual, np.zeros(len(u)), jac=dual.grad, method="BFGS")
pdf = dual.dist(res.x)

rv = numpy.random.choice(x, size=1000, p = pdf)

fig = sm.qqplot(rv, stats.norm, fit=True, line='45')
#stats.probplot(rv, dist="norm", plot=pylab)
pylab.show()

print "Mean: ", mean(rv), "   Var: ", var(rv)
```

#### 2.

However, as the following Q-Q plot shows, if we narrow the span to a narrow range, say [-1.5, 1.5], the distribution is no longer normal.   

The reason is that we should always choose a considerably large sampling range to approximate the real distribution.  

For example, if the sampling range is olny [-1.5, 1.5], since $ \Phi(1.5) - \Phi(-1.5) \approx 0.866 $, we would fail to capture the rest 0.134.

```python
bound = 1.5
x = np.arange(-bound, bound, .001)
a = np.array([x, x*x])
u = np.array([0., 1.])
e = np.array([0., 0.])
q = np.ones(np.size(x))/len(x)
dual = me.MaxEntDual(q, a, u, e)

res = minimize(dual.dual, np.zeros(len(u)), jac=dual.grad, method="BFGS")
pdf = dual.dist(res.x)

rv = numpy.random.choice(x, size=10000, p = pdf)

#stats.probplot(rv, dist="norm", plot=pylab)
fig = sm.qqplot(rv, stats.norm, fit=True, line='45')
pylab.show()
```

#### 3.
If we sample y from equal space, the result is not lognormal, the following plot is an example.

```python
#lim = 6
#lb, ub = np.exp(-lim), np.exp(lim)
#y = np.arange(lb, ub, .01)
#y = np.logspace(-150.0, 150.0, num=10000)
y = np.arange(1e-7, 10, .01)
a = np.array([np.log(y), np.log(y)*np.log(y)])
u = np.array([0., 1.])
e = np.array([0., 0.])
q = np.ones(np.size(y))/len(y)
#q = stats.norm.pdf(y, loc=0, scale=1)
dual = me.MaxEntDual(q, a, u, e)

res = minimize(dual.dual, np.zeros(len(u)), jac=dual.grad, method="BFGS")
pdf = dual.dist(res.x)

rv = numpy.random.choice(y, size=10000, p = pdf)
fig = sm.qqplot(rv, stats.lognorm, fit=True, line='45')
pylab.show()
```

#### 4.

The reason is that the equally spaced sample failed to capture the logarithm nature of the x-axis.

If we plot the pdf of $log(y)$, we can see that it is  centered at $x=-1$, but actually it should centered at $x = 0$.

```python
plot(np.log(y), pdf)
```

Often times, we run into conflicting constraints in practice,

2. show that you can find a solution by introducing error bounds, and shift the distribution to skew left or right by adjusting the error bounds related to these donstraints.

This exercise illustrates that error bounds improve the robustness of the ME method.

## Solution:
#### 1.
The following codes show that the optimizer failed to find a solution if we put confilcting constraints:

```python
x = np.arange(-10, 10, .001)
a = np.array([x*x*x, x*x*x])
u = np.array([-2., 2.])
e = np.array([0., 0.])
q = stats.norm.pdf(x, loc=0, scale=1)

dual = me.MaxEntDual(q, a, u, e)

res = minimize(dual.dual, np.zeros(len(u)), jac=dual.grad, method="BFGS")
res
```

#### 2.
We can introduce the error bound to meet either constraint.

For example, if we choose the first one:

```python
x = np.arange(-10, 10, .001)
a = np.array([x*x*x, x*x*x])
u = np.array([-2., 2.])
e = np.array([0.0, 0.1])
q = stats.norm.pdf(x, loc=0, scale=1)

dual = me.MaxEntDual(q, a, u, e)

res = minimize(dual.dual, np.zeros(len(u)), jac=dual.grad, method="BFGS")
print res

pdf = dual.dist(res.x)


#plot(x, pdf)

print "\n\n Skewness E[x^3] = ", sum(x*x*x*pdf)
```

If we choose to satisfy the second one, then:

```python
x = np.arange(-10, 10, .001)
a = np.array([x*x*x, x*x*x])
u = np.array([-2., 2.])
e = np.array([0.1, 0.0])
q = stats.norm.pdf(x, loc=0, scale=1)

dual = me.MaxEntDual(q, a, u, e)

res = minimize(dual.dual, np.zeros(len(u)), jac=dual.grad, method="BFGS")
print res

pdf = dual.dist(res.x)


#plot(x, pdf)

print "\n\n Skewness E[x^3] = ", sum(x*x*x*pdf)
```

#### 5.
#### A.

Suppose $A_1 = l, A_2 = l, A_3 = l, ..., A_{n} = l, A_(n+1) = r$

Then under Shapley, the allocation of $A_{n+1} = \frac{(n+1)! - (n)!}{(n+1)!} = \frac{n}{n+1}$

While the allocation of $A_1, A_2, ... , A_{n} = \frac{1-\frac{n}{n+1}}{n} = \frac{1}{n(n+1)}$

#### B.
If all left hands form a team, the situation under C-Shapley would be:

Allocation of $A_n = \frac{(n-1)!}{2(n-1)!} = \frac{1}{2}$

Allocation of $A_1, A_2, ..., A_{n-1} = 1-\frac{1}{2} = \frac{1}{2}$
